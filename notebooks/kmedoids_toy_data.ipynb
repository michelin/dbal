# KMedoids Query on toy data
In this example, we generate toy data, split between source and target domain.  
We then apply KMedoids query to select an optimal set of points from the target domain, to be used to train a ML model.

```python
from dbal.utils import toy_example
from dbal.query_methods import KMedoidsQuery

import numpy as np
from matplotlib import pyplot as plt
```

## Generate toy data

```python
np.random.seed(2)
Xs, Xt, f = toy_example()
ys = f(Xs)
yt = f(Xt)
```

```python
plt.scatter(Xs[:,0], Xs[:,1], s=3, label='Source domain')
plt.scatter(Xt[:,0], Xt[:,1], s=3, label='Target domain')
plt.legend()
```

## Apply KMedoids Query

```python
n_queries = 10

np.random.seed(0)
kmedoids = KMedoidsQuery()
queries = kmedoids.fit_predict(Xt, Xs, n_queries=n_queries)
```

```python
plt.scatter(Xs[:,0], Xs[:,1], s=3, label='Source domain')
plt.scatter(Xt[:,0], Xt[:,1], s=3, label='Target domain')
plt.scatter(Xt[queries,0], Xt[queries,1], marker='o', s=50, label="Queries", edgecolor=(0,0,0,1), facecolor=(0,0,0,0.5))
plt.legend()
```
